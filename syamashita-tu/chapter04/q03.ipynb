$f(x)=x^2$

$f(1)=1^2=1$

$f'(x)=2x$

$f'(1)=2$

\begin{align}
y&=f(1)+f'(1)(x-1)\\
&=1+2(x-1)\\
&=2x-1
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def tangent(x):
    return 2*x - 1

x = np.linspace(-2, 3, 400)

plt.figure(figsize=(7, 5))
plt.plot(x, f(x), label=r"$f(x)=x^2$")
plt.plot(x, tangent(x), label=r"tangent $y=2x-1$")
plt.scatter([1], [f(1)], label="tangent point")
plt.axhline(0, linewidth=0.8)
plt.axvline(0, linewidth=0.8)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Tangent line at x=1")
plt.grid(True)
plt.legend()
plt.show()

print("f(1) =", f(1))
print("f'(1) =", 2)
print("接線：y = 2x - 1")
```

$y=2x-1$は，点 $(1,1)$ で関数 $f(x)=x^2$と接する．
この点の近くでは，曲線と接線の値が近くなるため，接線は関数を局所的に直線で近似しているといえる．
