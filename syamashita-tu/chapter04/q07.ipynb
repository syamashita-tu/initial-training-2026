\begin{align}
v&=
\begin{pmatrix}
1\\
0
\end{pmatrix},\quad
R&=
\begin{pmatrix}
\cos\theta & -\sin\theta\\
\sin\theta & \cos\theta
\end{pmatrix}
\end{align}

$\cos 90^\circ=0,\quad \sin 90^\circ=1$

\begin{align}
R=
\begin{pmatrix}
0 & -1\\
1 & 0
\end{pmatrix}
\end{align}

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def rotation_matrix(degree):
    theta = np.deg2rad(degree)
    return np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta),  np.cos(theta)]])

v = np.array([1, 0])
degrees = [0, 30, 60, 90, 180]

rows = []
points = []

for deg in degrees:
    R = rotation_matrix(deg)
    rotated = R @ v
    points.append(rotated)
    rows.append({
        "degree": deg,
        "x": rotated[0],
        "y": rotated[1],
        "length": np.linalg.norm(rotated)
    })

display(pd.DataFrame(rows))

points = np.array(points)

plt.figure(figsize=(6, 6))
for deg, p in zip(degrees, points):
    plt.quiver(0, 0, p[0], p[1], angles="xy", scale_units="xy", scale=1)
    plt.text(p[0], p[1], f"{deg}°")

circle_theta = np.linspace(0, 2*np.pi, 400)
plt.plot(np.cos(circle_theta), np.sin(circle_theta), linestyle="--", label="unit circle")

plt.xlim(-1.2, 1.2)
plt.ylim(-1.2, 1.2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Rotation of (1,0)")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.legend()
plt.show()
```

回転後のベクトルはいずれも長さが $1$ に近い．したがって，回転行列はベクトルの長さを保ったまま，向きだけを変える変換．
