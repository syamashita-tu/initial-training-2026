```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

rng = np.random.default_rng(0)
n_trials_list = [10, 100, 1000, 10000]

rows = []

for n_trials in n_trials_list:
    rolls = rng.integers(1, 7, size=n_trials)
    even_ratio = np.mean(rolls % 2 == 0)
    rows.append({
        "試行回数": n_trials,
        "偶数の割合": even_ratio,
        "理論値との差": even_ratio - 0.5
    })

df = pd.DataFrame(rows)
display(df)

plt.figure(figsize=(7, 5))
plt.plot(df["試行回数"], df["偶数の割合"], marker="o", label="simulation")
plt.axhline(0.5, linestyle="--", label="theoretical probability")
plt.xscale("log")
plt.xlabel("number of trials")
plt.ylabel("ratio of even numbers")
plt.title("Dice simulation")
plt.grid(True)
plt.legend()
plt.show()
```

試行回数が少ないと，偶数が出る割合は理論値 $0.5$ からずれることがある．しかし，試行回数を増やすと，偶数が出る割合は $0.5$ に近づきやすい．これは大数の法則の例．
