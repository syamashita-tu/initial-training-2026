\begin{align}
A&=
\begin{pmatrix}
2 & 0\\
0 & 1
\end{pmatrix},\quad
v&=
\begin{pmatrix}
1\\
2
\end{pmatrix}
\end{align}

\begin{align}
Av=
\begin{pmatrix}
2 & 0\\
0 & 1
\end{pmatrix}
\begin{pmatrix}
1\\
2
\end{pmatrix}
=
\begin{pmatrix}
2\cdot 1+0\cdot 2\\
0\cdot 1+1\cdot 2
\end{pmatrix}
=
\begin{pmatrix}
2\\
2
\end{pmatrix}
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 0],
              [0, 1]])
v = np.array([1, 2])
Av = A @ v

print("A =")
print(A)
print("v =", v)
print("Av =", Av)

plt.figure(figsize=(6, 6))
origin = np.array([[0, 0], [0, 0]])
vectors = np.array([v, Av])

plt.quiver(origin[:, 0], origin[:, 1], vectors[:, 0], vectors[:, 1],
           angles="xy", scale_units="xy", scale=1)
plt.text(v[0], v[1], "v")
plt.text(Av[0], Av[1], "Av")
plt.xlim(0, 3)
plt.ylim(0, 3)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Matrix transformation")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
```

この行列は，$x$ 成分を $2$ 倍し，$$y$ 成分はそのままにする．
\begin{align}
\begin{pmatrix}
1\\
2
\end{pmatrix}
\mapsto
\begin{pmatrix}
2\\
2
\end{pmatrix}
\end{align}
