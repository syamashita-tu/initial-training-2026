\begin{align}
A=
\begin{pmatrix}
2 & 0\\
0 & 3
\end{pmatrix}
\end{align}

\begin{align}
e_1=
\begin{pmatrix}
1\\
0
\end{pmatrix},
\quad
e_2=
\begin{pmatrix}
0\\
1
\end{pmatrix}
\end{align}

\begin{align}
Ae_1=
\begin{pmatrix}
2 & 0\\
0 & 3
\end{pmatrix}
\begin{pmatrix}
1\\
0
\end{pmatrix}
=
\begin{pmatrix}
2\\
0
\end{pmatrix}
=
2e_1
\end{align}

\begin{align}
Ae_2=
\begin{pmatrix}
2 & 0\\
0 & 3
\end{pmatrix}
\begin{pmatrix}
0\\
1
\end{pmatrix}
=
\begin{pmatrix}
0\\
3
\end{pmatrix}
=
3e_2
\end{align}

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([[2, 0],
              [0, 3]])

eigvals, eigvecs = np.linalg.eig(A)

print("固有値")
print(eigvals)
print("固有ベクトル")
print(eigvecs)

vectors = [
    np.array([1, 0]),
    np.array([0, 1]),
    np.array([1, 1]),
    np.array([2, 1])
]

plt.figure(figsize=(7, 7))

for v in vectors:
    Av = A @ v
    plt.quiver(0, 0, v[0], v[1], angles="xy", scale_units="xy", scale=1)
    plt.quiver(0, 0, Av[0], Av[1], angles="xy", scale_units="xy", scale=1, alpha=0.55)
    plt.text(v[0], v[1], f"v={tuple(v)}")
    plt.text(Av[0], Av[1], f"Av={tuple(Av)}")

plt.xlim(-0.5, 5)
plt.ylim(-0.5, 5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Original vectors and transformed vectors")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()

def angle_degree(v):
    return np.rad2deg(np.arctan2(v[1], v[0]))

for v in vectors:
    Av = A @ v
    print(f"v={v}, Av={Av}, angle(v)={angle_degree(v):.2f}, angle(Av)={angle_degree(Av):.2f}")
```

$(1,0)^T$ と $(0,1)^T$ は，行列をかけても向きが変わらない．これは，変換後が元のベクトルの定数倍になるため．

一方で，

$$
A
\begin{pmatrix}
1 \\
1
\end{pmatrix}
=
\begin{pmatrix}
2 \\
3
\end{pmatrix}
$$

これは $(1,1)^T$ の定数倍ではないため，一般には向きが変わる．
