$ x=[1,2,3,4,5] $

$\bar{x}=\frac{1+2+3+4+5}{5}=3$

$\frac{1}{5}\sum_{i=1}^{5}(x_i-\bar{x})^2$

\begin{align}
\frac{(1-3)^2+(2-3)^2+(3-3)^2+(4-3)^2+(5-3)^2}{5}
=
\frac{4+1+0+1+4}{5}
=2
\end{align}

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4, 5])

mean_x = np.mean(x)
var_x = np.var(x)

print("データ x =", x)
print("平均 =", mean_x)
print("分散 =", var_x)
```

```python
rng = np.random.default_rng(0)
n = 10000

samples = rng.normal(loc=0, scale=1, size=n)

sample_mean = np.mean(samples)
sample_var = np.var(samples)

print("乱数の個数 =", n)
print("標本平均 =", sample_mean)
print("標本分散 =", sample_var)

plt.figure(figsize=(7, 5))
plt.hist(samples, bins=50, density=True, alpha=0.7)
plt.xlabel("x")
plt.ylabel("density")
plt.title("Histogram of standard normal samples")
plt.grid(True)
plt.show()
```

```python
n_list = [10, 100, 1000, 10000, 100000]
rows = []

rng = np.random.default_rng(1)

for n in n_list:
    samples = rng.normal(loc=0, scale=1, size=n)
    rows.append({
        "データ数": n,
        "標本平均": np.mean(samples),
        "標本分散": np.var(samples),
        "平均の理論値との差": np.mean(samples) - 0,
        "分散の理論値との差": np.var(samples) - 1
    })

df = pd.DataFrame(rows)
display(df)

plt.figure(figsize=(7, 5))
plt.plot(df["データ数"], df["標本平均"], marker="o", label="sample mean")
plt.axhline(0, linestyle="--", label="theoretical mean")
plt.xscale("log")
plt.xlabel("number of samples")
plt.ylabel("mean")
plt.title("Sample mean")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(7, 5))
plt.plot(df["データ数"], df["標本分散"], marker="o", label="sample variance")
plt.axhline(1, linestyle="--", label="theoretical variance")
plt.xscale("log")
plt.xlabel("number of samples")
plt.ylabel("variance")
plt.title("Sample variance")
plt.grid(True)
plt.legend()
plt.show()
```

標準正規分布の理論的な平均は 0，分散は 1．生成した乱数から求めた標本平均と標本分散は，乱数のため完全には一致しない．しかし，データ数を増やすと，平均は 0，分散は 1 に近づきやすい．
