\begin{align}
a&=
\begin{pmatrix}
3\\
4
\end{pmatrix},
\quad
b&=
\begin{pmatrix}
1\\
2
\end{pmatrix}
\end{align}

$\|a\|=\sqrt{3^2+4^2}=\sqrt{25}=5$

$\|b\|=\sqrt{1^2+2^2}=\sqrt{5}$

$a\cdot b=3\cdot 1+4\cdot 2=11$

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.array([3, 4])
b = np.array([1, 2])

norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
dot_ab = np.dot(a, b)

print("||a|| =", norm_a)
print("||b|| =", norm_b)
print("a・b =", dot_ab)

plt.figure(figsize=(6, 6))
origin = np.array([[0, 0], [0, 0]])
vectors = np.array([a, b])

plt.quiver(origin[:, 0], origin[:, 1], vectors[:, 0], vectors[:, 1],
           angles="xy", scale_units="xy", scale=1)
plt.text(a[0], a[1], "a")
plt.text(b[0], b[1], "b")
plt.xlim(0, 4)
plt.ylim(0, 5)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Vectors a and b")
plt.grid(True)
plt.gca().set_aspect("equal", adjustable="box")
plt.show()
```

の長さは，直角三角形の辺の長さ $3,4,5$ と対応．三平方の定理から $\|a\|=5$．
