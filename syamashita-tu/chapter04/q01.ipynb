$f(x)=x^2$

\begin{aligned}
f(-2)&=(-2)^2=4\\
f(-1)&=(-1)^2=1\\
f(0)&=0^2=0\\
f(1)&=1^2=1\\
f(2)&=2^2=4
\end{aligned}

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x = np.linspace(-3, 3, 400)
y = f(x)

manual_x = np.array([-2, -1, 0, 1, 2])
manual_y = f(manual_x)

plt.figure(figsize=(7, 5))
plt.plot(x, y, label=r"$f(x)=x^2$")
plt.scatter(manual_x, manual_y, label="manual points")
plt.axhline(0, linewidth=0.8)
plt.axvline(0, linewidth=0.8)
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Graph of f(x)=x^2")
plt.grid(True)
plt.legend()
plt.show()

print("手計算で求めた点")
for xi, yi in zip(manual_x, manual_y):
    print(f"f({xi}) = {yi}")
```

$x=0$ の近くでは関数値が最小となり，$x$ が正負どちらに少し動いても $f(x)$ は増加．したがって，グラフは $x=0$ の近くで下に凸に曲がる形．
