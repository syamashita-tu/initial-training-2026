連続時間信号 $x(t) = \cos(4 \pi t)$ をサンプリング周波数 $F_s = 16$ Hz で標本化した離散時間信号 $x[n]$ の16点 DFT を計算し，
横軸を周波数$F$ [Hz] として振幅スペクトルと位相スペクトルをプロットしなさい．
また，スペクトルが2本立つ理由を，オイラーの公式およびDFTの周期性を踏まえて考察しなさい．

$$
\cos(4\pi t)
=
\frac{1}{2}e^{j4\pi t}
+
\frac{1}{2}e^{-j4\pi t}
$$

したがって，$+2\ \mathrm{Hz}$ と $-2\ \mathrm{Hz}$ の 2 本のスペクトル．

DFT の周期性により，$-2\ \mathrm{Hz}$ は

$$
k = N - 2 = 14
$$

に現れる．

$$
|X[2]| = 8,\qquad |X[14]| = 8
$$

```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 16
N = 16
n = np.arange(N)
t = n / Fs

x = np.cos(4 * np.pi * t)
X = np.fft.fft(x)
F = np.fft.fftfreq(N, d=1/Fs)

amp = np.abs(X)
phase = np.angle(X)

plt.figure()
plt.stem(F, amp)
plt.xlabel("F [Hz]")
plt.ylabel("|X[k]|")
plt.title("Amplitude spectrum")
plt.grid()
plt.show()

plt.figure()
plt.stem(F, phase)
plt.xlabel("F [Hz]")
plt.ylabel("phase [rad]")
plt.title("Phase spectrum")
plt.grid()
plt.show()

for i in np.where(amp > 1e-8)[0]:
    print(i, F[i], amp[i], phase[i])
```

- 余弦波は正周波数と負周波数の和
- そのため，振幅スペクトルは 2 本
- DFT は周期的なので，負周波数 $-2\ \mathrm{Hz}$ は $k=14$ に対応
