$N = 4$ のとき，DFT行列 $\boldsymbol{F}_N$ を手で求めなさい．
また， $\boldsymbol{F}_N$ を使って，$x[n] = (1, 1, 0, 0)^\top$ の DFT を手計算しなさい．
得られた $X[k]$ を， np.fft.fft を使ってDFTを計算した結果と比較しなさい．

```python

N = 4

x = np.array([1, 1, 0, 0])
for k in range(N):
    s = 0
    for l in range(N):
        s += x[l] * np.exp(-2j * np.pi * k * l / N)
    print(s)
    
```

```python
import numpy as np

x = np.array([1, 1, 0, 0])
X_fft = np.fft.fft(x)

print(X_fft)
```
