$N = 8$ のとき，複素正弦波信号 $\phi_k[n] = e^{j 2 \pi k n / N}, \quad k = 0, 1, \cdots, N-1$ 間の内積
$\langle \phi_k, \phi_l \rangle$ を計算し， $\langle \phi_k, \phi_l \rangle$ を $k$ 行 $l$ 列の要素とする行列を作りなさい．
この行列から $\phi_0, \cdots, \phi_{N-1}$ が互いに直交していることを確認しなさい．

$$
\phi_k[n] = e^{j2\pi kn/N}
$$

内積は

$$
\left\langle \phi_k, \phi_l \right\rangle
=
\sum_{n=0}^{N-1}
\overline{\phi_k[n]}\phi_l[n]
$$

$N=8$ のとき，

$$
\left\langle \phi_k, \phi_l \right\rangle
=
\begin{cases}
8 & k=l \\
0 & k\ne l
\end{cases}
$$

したがって，行列は

$$
\begin{bmatrix}
8 & 0 & 0 & 0 & 0 & 0 & 0 & 0 \\
0 & 8 & 0 & 0 & 0 & 0 & 0 & 0 \\
0 & 0 & 8 & 0 & 0 & 0 & 0 & 0 \\
0 & 0 & 0 & 8 & 0 & 0 & 0 & 0 \\
0 & 0 & 0 & 0 & 8 & 0 & 0 & 0 \\
0 & 0 & 0 & 0 & 0 & 8 & 0 & 0 \\
0 & 0 & 0 & 0 & 0 & 0 & 8 & 0 \\
0 & 0 & 0 & 0 & 0 & 0 & 0 & 8
\end{bmatrix}
$$

```python
import numpy as np

N = 8
n = np.arange(N)
k = np.arange(N)[:, None]

phi = np.exp(1j * 2 * np.pi * k * n / N)
G = phi.conj() @ phi.T

print(np.round(G, 10))
```
