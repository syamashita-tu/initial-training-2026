連続時間信号 $x(t) = e^{j 4 \pi t}$ をサンプリング周波数 $F_s = 16$ Hz で標本化した離散時間信号 $x[n]$ の16点 DFT を計算し，
横軸を周波数インデックス$k$として振幅スペクトルをプロットしなさい．
また，横軸を周波数$F$ [Hz] とした振幅スペクトルもプロットしなさい．

$$
x(t) = e^{j4\pi t} = e^{j2\pi \cdot 2t}
$$

なので，周波数は

$$
F = 2\ \mathrm{Hz}
$$

16 点 DFT では

$$
k = \frac{FN}{F_s}
=
\frac{2 \cdot 16}{16}
=
2
$$

したがって，振幅スペクトルは $k=2$ にピーク，

$$
|X[2]| = 16
$$

```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 16
N = 16
n = np.arange(N)
t = n / Fs

x = np.exp(1j * 4 * np.pi * t)
X = np.fft.fft(x)

k = np.arange(N)
F = np.fft.fftfreq(N, d=1/Fs)

plt.figure()
plt.stem(k, np.abs(X))
plt.xlabel("k")
plt.ylabel("|X[k]|")
plt.title("Amplitude spectrum vs index")
plt.grid()
plt.show()

plt.figure()
plt.stem(F, np.abs(X))
plt.xlabel("F [Hz]")
plt.ylabel("|X[k]|")
plt.title("Amplitude spectrum vs frequency")
plt.grid()
plt.show()

print("peak index:", np.argmax(np.abs(X)))
print("peak frequency:", F[np.argmax(np.abs(X))])
print("peak amplitude:", np.max(np.abs(X)))
```
