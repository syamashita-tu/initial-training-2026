信号に含まれる所望の信号のみを通過し，それ以外を阻止するシステムを，フィルタ (filter) と呼ぶ．
特に，低周波成分を通過し高周波成分を阻止するフィルタを 低域通過フィルタ (low-pass filter) と呼び，
高周波成分を通過し低周波成分を阻止するフィルタを 高域通過フィルタ (high-pass filter) と呼ぶ．
以下のインパルス応答 $h = (h[0], h[1], \cdots)^\top$ を持つシステム $\mathcal{H}$ が
低域通過フィルタであるか高域通過フィルタであるかを，
$h$ の振幅スペクトルをプロットして考察しなさい．

1. $h = (1/3, 1/3, 1/3)^\top$
2. $h = (1, -1)^\top$

周波数応答の振幅 $|H[k]|$ が低周波で大きく高周波で小さい場合は低域通過フィルタです．反対に，低周波で小さく高周波で大きい場合は高域通過フィルタです．

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=6, suppress=True)


h1 = np.array([1/3, 1/3, 1/3], dtype=float)
h2 = np.array([1, -1], dtype=float)

Nfft = 4096
freq = np.fft.rfftfreq(Nfft, d=1.0)

H1 = np.fft.rfft(h1, n=Nfft)
H2 = np.fft.rfft(h2, n=Nfft)

plt.figure()
plt.plot(freq, np.abs(H1))
plt.xlabel("normalized frequency [cycles/sample]")
plt.ylabel("|H1[k]|")
plt.title("frequency response of h1 = [1/3, 1/3, 1/3]")
plt.grid(True)
plt.show()

plt.figure()
plt.plot(freq, np.abs(H2))
plt.xlabel("normalized frequency [cycles/sample]")
plt.ylabel("|H2[k]|")
plt.title("frequency response of h2 = [1, -1]")
plt.grid(True)
plt.show()

print("h1: |H(0)| =", abs(H1[0]), ", |H(π)| =", abs(H1[-1]))
print("h2: |H(0)| =", abs(H2[0]), ", |H(π)| =", abs(H2[-1]))
```

- $h_1=(1/3,1/3,1/3)^\top$ は3点移動平均です．低周波では振幅が大きく，高周波では振幅が小さくなるため，低域通過フィルタです．
- $h_2=(1,-1)^\top$ は1サンプル差分です．直流成分では $|H(0)|=0$ となり，高周波側で振幅が大きくなるため，高域通過フィルタです．
