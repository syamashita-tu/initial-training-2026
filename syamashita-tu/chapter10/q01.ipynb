以下のシステム $y = \mathcal{S}\{x\}$ の線形性と時不変性を確認し，線形時不変システムをすべて挙げなさい．
\begin{align*}
y[n] &= x[n] + 1　\\
y[n] &= 2 x[n] + x[n-1]\\
y[n] &= x[-n]\\
y[n] &= x[n]^2\\
y[n] &= 0.9 y[n-1] + x[n]\\
\end{align*}

線形性：$\mathcal{S}\{a x_1 + b x_2\}=a\mathcal{S}\{x_1\}+b\mathcal{S}\{x_2\}$

時不変性：$\mathcal{S}\{\mathcal{D}_{n_0}\{x\}\}=\mathcal{D}_{n_0}\{\mathcal{S}\{x\}\}$

| 番号 | システム | 線形性 | 時不変性 | LTI |
|---:|---|---|---|---|
| 1 | $y[n]=x[n]+1$ | なし | あり | なし |
| 2 | $y[n]=2x[n]+x[n-1]$ | あり | あり | あり |
| 3 | $y[n]=x[-n]$ | あり | なし | なし |
| 4 | $y[n]=x[n]^2$ | なし | あり | なし |
| 5 | $y[n]=0.9y[n-1]+x[n]$ | あり | あり | あり |

## 各システムの理由

### 1．$y[n]=x[n]+1$

零入力 $x[n]=0$ に対して出力が $1$ になるため，線形性なし．入力を遅らせても，出力を遅らせても，どちらも $x[n-n_0]+1$ になるため，時不変．

### 2．$y[n]=2x[n]+x[n-1]$

入力の定数倍と和に対して出力も同じ定数倍と和になる．また，係数と遅延量が時刻 $n$ に依存しないため，時不変．

### 3．$y[n]=x[-n]$

時間反転は線形演算．しかし，入力を $n_0$ 遅延すると $x[-n-n_0]$，出力を $n_0$ 遅延すると $x[-n+n_0]$ になり，一般には一致しないため時不変ではない．

### 4．$y[n]=x[n]^2$

二乗により交差項が生じるため，線形性はなし．一方で各時刻ごとの二乗なので，入力遅延と出力遅延は一致し，時不変．

### 5．$y[n]=0.9y[n-1]+x[n]$

初期条件をゼロとする通常の信号処理システムとして扱うと，過去出力と入力の線形結合で構成されるため線形．係数 $0.9$ が時刻に依存しないため時不変．したがってLTI．
