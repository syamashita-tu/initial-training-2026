1の各システムのインパルス応答を求めなさい．また，インパルス応答の長さを求めなさい．

\begin{equation}
\delta[n]=
\begin{cases}
1 & n=0\\
0 & n\neq 0
\end{cases}
\end{equation}

```python
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(precision=6, suppress=True)


n = np.arange(-5, 16)

delta = (n == 0).astype(float)
u = (n >= 0).astype(float)

h1 = delta + 1
h2 = 2*(n == 0).astype(float) + (n == 1).astype(float)
h3 = delta
h4 = delta
h5 = (0.9**n) * u
h5[n < 0] = 0

responses = [
    ("1: δ[n]+1", h1),
    ("2: 2δ[n]+δ[n-1]", h2),
    ("3: δ[n]", h3),
    ("4: δ[n]", h4),
    ("5: 0.9^n u[n]", h5),
]

for title, h in responses:
    plt.figure()
    plt.stem(n, h)
    plt.xlabel("n")
    plt.ylabel("h[n]")
    plt.title(title)
    plt.grid(True)
    plt.show()
```

各システムにインパルス入力 $x[n]=\delta[n]$ を入れたときの出力は次の通り．

| 番号 | システム | インパルス入力に対する出力 | 長さ |
|---:|---|---|---|
| 1 | $y[n]=x[n]+1$ | $y[n]=\delta[n]+1$ | 無限長 |
| 2 | $y[n]=2x[n]+x[n-1]$ | $h[n]=2\delta[n]+\delta[n-1]$ | 2 |
| 3 | $y[n]=x[-n]$ | $y[n]=\delta[-n]=\delta[n]$ | 1 |
| 4 | $y[n]=x[n]^2$ | $y[n]=\delta[n]^2=\delta[n]$ | 1 |
| 5 | $y[n]=0.9y[n-1]+x[n]$ | $h[n]=0.9^n u[n]$ | 無限長 |

5番ではゼロ初期条件を仮定．このとき，$n=0$ で $h[0]=1$，以後 $h[n]=0.9h[n-1]$ となるため，指数的に減衰する無限長応答．
