大きさ $128 \times 128$ の2次元正弦波 $x[n_1, n_2] = \cos(\omega_1 n_1 + \omega_2 n_2)$ を，
以下の $(\omega_1, \omega_2)$ の組み合わせについて生成し，並べて表示しなさい
（```np.meshgrid``` を使うとよい）．
ただし， ```cmap='gray'``` とすること．
その後，縞模様の向きおよび細かさと $(\omega_1, \omega_2)$ の関係を考察しなさい．

- $(\omega_1, \omega_2) = (\pi/8, 0)$
- $(\omega_1, \omega_2) = (0, \pi/8)$
- $(\omega_1, \omega_2) = (\pi/8, \pi/8)$
- $(\omega_1, \omega_2) = (\pi/2, \pi/2)$





````python
import numpy as np
import matplotlib.pyplot as plt

N = 128

n1 = np.arange(N)
n2 = np.arange(N)

N1, N2 = np.meshgrid(n1, n2, indexing="ij")

params = [
    (np.pi / 8, 0, r"$(\omega_1,\omega_2)=(\pi/8,0)$"),
    (0, np.pi / 8, r"$(\omega_1,\omega_2)=(0,\pi/8)$"),
    (np.pi / 8, np.pi / 8, r"$(\omega_1,\omega_2)=(\pi/8,\pi/8)$"),
    (np.pi / 2, np.pi / 2, r"$(\omega_1,\omega_2)=(\pi/2,\pi/2)$"),
]

fig, ax = plt.subplots(1, 4, figsize=(14, 4))

for i, (w1, w2, title) in enumerate(params):
    x = np.cos(w1 * N1 + w2 * N2)

    ax[i].imshow(x, cmap="gray", vmin=-1, vmax=1)
    ax[i].set_title(title)
    ax[i].axis("off")

plt.tight_layout()
plt.show()
````

- 2次元正弦波の縞模様の向きは，角周波数ベクトル $(\omega 1,\omega 2)$ によって決まる．
- $\omega1$ のみが非零の場合は $n1$ 方向に値が変化するため横縞となり，$\omega2$ のみが非零の場合は $n2$ 方向に値が変化するため縦縞となる．
- また，$\omega1$ と $\omega2$ がともに非零の場合は，両方向に値が変化するため斜め縞となる．
- さらに，$\omega1$，$\omega2$ の値が大きいほど空間周期は短くなり，縞模様は細かくなる．
