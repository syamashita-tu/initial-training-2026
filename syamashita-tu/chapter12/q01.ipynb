画像 ```cat.png``` を ```Image.open``` で読み込み，NumPy 配列に変換して， 配列の形状 (shape)，データ型 (dtype)，画素値の最小値・最大値を表示しなさい． また，```img.convert('L')``` でグレースケール画像に変換し，同様に形状・データ型・最小値・最大値を確認しなさい． カラー画像とグレースケール画像で形状がどのように異なるか答えなさい．

````python
from PIL import Image 
import matplotlib.pyplot as plt
import numpy as np

image = Image.open('/home/syamashita/initial-training-2026/syamashita-tu/chapter12/cat.png')
image_array = np.array(image)
print(f"Shape: {image_array.shape}")
print(f"Dtype: {image_array.dtype}")
print(f"Max: {np.max(image_array)}")
print(f"Min: {np.min(image_array)}")

image_gray = image.convert('L')
image_gray_array = np.array(image_gray)
print(f"Gray Shape: {image_gray_array.shape}")
print(f"Gray Dtype: {image_gray_array.dtype}")
print(f"Gray Max: {np.max(image_gray_array)}")
print(f"Gray Min: {np.min(image_gray_array)}")
````
