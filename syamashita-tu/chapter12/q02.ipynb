1で読み込んだRGB画像の配列から，赤・緑・青の各チャネル ```arr[:, :, 0]```, ```arr[:, :, 1]```, ```arr[:, :, 2]``` を取り出し， 元のカラー画像と合わせて4枚を ```cmap='gray'``` で並べて表示しなさい（plt.subplot を使うとよい）． 全チャネルで共通して明るい部分や，あるチャネルだけが明るい部分などがあるか観察しなさい．

````python
from PIL import Image 
import matplotlib.pyplot as plt
import numpy as np

image = Image.open('/home/syamashita/initial-training-2026/syamashita-tu/chapter12/cat.png')
image_array = np.array(image)

R_array = image_array[:, :, 0]
G_array = image_array[:, :, 1]
B_array = image_array[:, :, 2]

fig, ax = plt.subplots(1, 4, figsize=(16, 4))

ax[0].imshow(image_array)
ax[0].set_title("Original RGB")
ax[0].axis("off")

ax[1].imshow(R_array, cmap="gray", vmin=0, vmax=255)
ax[1].set_title("Red channel")
ax[1].axis("off")

ax[2].imshow(G_array, cmap="gray", vmin=0, vmax=255)
ax[2].set_title("Green channel")
ax[2].axis("off")

ax[3].imshow(B_array, cmap="gray", vmin=0, vmax=255)
ax[3].set_title("Blue channel")
ax[3].axis("off")

plt.tight_layout()
plt.show()
````
