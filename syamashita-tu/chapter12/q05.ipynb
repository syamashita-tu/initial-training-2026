グレースケール画像に対して以下の演算を施した画像を生成し，表示しなさい．

- ```np.roll``` を使って，$n_1$ 方向（縦）に 50 画素，$n_2$ 方向（横）に 100 画素シフトした画像
- スライス ```[::-1, :]``` を使って上下反転した画像
- スライス ```[:, ::-1]``` を使って左右反転した画像
- スライス ```[::2, ::2]``` を使って縦横それぞれ 1/2 に間引いた（縮小した）画像

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("/home/syamashita/initial-training-2026/syamashita-tu/chapter12/cat.png").convert("L")
img_array = np.array(img)

a = np.roll(img_array, shift=[50, 0], axis=0)
b = img_array[::-1, :]
c = img_array[:, ::-1]
d = img_array[::2, ::2]

fig, ax = plt.subplots(1, 4, figsize=(12, 3))
ax[0].imshow(a, cmap="gray")
ax[0].set_title("Shifted")
ax[1].set_title("Flipped Vertically")
ax[1].imshow(b, cmap="gray")
ax[2].set_title("Flipped Horizontally")
ax[2].imshow(c, cmap="gray")
ax[3].set_title("Subsampled")
ax[3].imshow(d, cmap="gray")
ax[0].axis("off")
ax[1].axis("off")
ax[2].axis("off")
ax[3].axis("off")
plt.show()
````
