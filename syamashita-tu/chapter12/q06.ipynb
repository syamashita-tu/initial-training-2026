回転角 $\theta = 90^\circ$ のときの回転行列 $\boldsymbol{R}(\theta)$ を手計算しなさい．
また，座標 $(n_1, n_2) = (2, 1)$ を $\theta = 90^\circ$ 回転した後の座標を手計算しなさい．
さらに，Pillow の ```img.rotate(90)``` を使って画像を $90^\circ$ 回転した結果を表示しなさい．

\begin{align*}
R(\theta)
&=
\begin{pmatrix}
\cos\theta & -\sin\theta \\
\sin\theta & \cos\theta
\end{pmatrix}
\\
R(90^\circ)
&=
\begin{pmatrix}
\cos 90^\circ & -\sin 90^\circ \\
\sin 90^\circ & \cos 90^\circ
\end{pmatrix}
\\
&=
\begin{pmatrix}
0 & -1 \\
1 & 0
\end{pmatrix}
\end{align*}

````python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open("/home/syamashita/initial-training-2026/syamashita-tu/chapter12/cat.png").convert("L")

theta = 90
img_array_rotated = img.rotate(theta)

plt.imshow(img_array_rotated, cmap="gray")
plt.axis("off")
plt.show()

````
