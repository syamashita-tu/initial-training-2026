2次元DFTが1次元DFTを行方向・列方向に順に適用したものと一致することを確認する．
適当な $8 \times 8$ の2次元信号に対して，```np.fft.fft``` を ```axis=1```（行方向），続いて ```axis=0```（列方向）に適用した結果と，
```np.fft.fft2``` の結果が一致することを ```np.allclose``` で確認しなさい．


````python
import numpy as np

# 8×8 の2次元信号
x = np.arange(64).reshape(8, 8).astype(float)

# 1次元DFTを行方向 axis=1 に適用
X_row = np.fft.fft(x, axis=1)

# 続いて列方向 axis=0 に適用
X_seq = np.fft.fft(X_row, axis=0)

# 2次元DFT
X_fft2 = np.fft.fft2(x)

# 一致確認
result = np.allclose(X_seq, X_fft2)

# 最大誤差
max_error = np.max(np.abs(X_seq - X_fft2))

print("x =")
print(x)

print("np.allclose(X_seq, X_fft2) =", result)
print("max error =", max_error)
````
