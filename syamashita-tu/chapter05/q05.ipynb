```python
def quantize(x, n_bits, x_min=-1.0, x_max=1.0):
    """振幅を [x_min, x_max] の範囲で n_bits 段階に量子化"""
    L = 2 ** n_bits
    step = (x_max - x_min) / (L - 1)
    return np.round((x - x_min) / step) * step + x_min
```

```python
import numpy as np
import matplotlib.pyplot as plt
 
Fs = 1000             # サンプリング周波数 [Hz]
Ts = 1.0 / Fs         # サンプリング周期 [秒]
duration = 1.0        # 信号長 [秒]
 
n = np.arange(int(duration * Fs))     # 0, 1, 2, ..., N-1
t = n * Ts                            # 各サンプルに対応する時刻

x = np.sin(10 * np.pi * t)        # x[n] = sin(2π f0 (n Ts))

de_x = quantize(x, n_bits=3)

plt.plot(t, x)
plt.plot(t, de_x)
plt.xlabel("Time [s]"); plt.ylabel("x[n]")
plt.show()
```
