```python
import numpy as np
import matplotlib.pyplot as plt
 
Fs_fast = 1000             # サンプリング周波数 [Hz]
Ts_fast = 1.0 / Fs_fast        # サンプリング周期 [秒]
Fs_slow = 8             # サンプリング周波数 [Hz]
Ts_slow = 1.0 / Fs_slow        # サンプリング周期 [秒]
duration = 1.0        # 信号長 [秒]


 
n_fast = np.arange(int(duration * Fs_fast))     # 0, 1, 2, ..., N-1
t_fast = n_fast * Ts_fast                            # 各サンプルに対応する時刻

n_slow = np.arange(int(duration * Fs_slow))     # 0, 1, 2, ..., N-1
t_slow = n_slow * Ts_slow                            # 各サンプルに対応する時

F = 1
x1_fast = np.cos(2 * np.pi * F * t_fast)
x1_slow = np.cos(2 * np.pi * F * t_slow)        # x[n] = sin(2π f0 (n Ts_fast))
x2_fast = np.cos(14 * np.pi * F * t_fast)
x2_slow = np.cos(14 * np.pi * F * t_slow)        # x[n] = sin(2π f0 (n Ts_fast))
 
plt.plot(t_fast, x1_fast, label='Fast Sampling')
plt.stem(t_slow, x1_slow, label='Slow Sampling')
plt.xlabel("Time [s]"); plt.ylabel("x[n]")
plt.show()
plt.plot(t_fast, x2_fast, label='Fast Sampling')
plt.stem(t_slow, x2_slow, label='Slow Sampling')
plt.xlabel("Time [s]"); plt.ylabel("x[n]")
plt.show()
```
