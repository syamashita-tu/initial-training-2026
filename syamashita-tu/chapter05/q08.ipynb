```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 1000
T = 2
t = np.arange(0, T, 1/Fs)

def x(d, t):
    return 4 / (np.pi * d) * np.sin(2 * np.pi * d * t)

signals = {
    r"$x_1(t)$": x(1, t),
    r"$x_1(t)+x_3(t)$": x(1, t) + x(3, t),
    r"$x_1(t)+x_3(t)+x_5(t)$": x(1, t) + x(3, t) + x(5, t),
    r"$\sum_{i=1}^{20} x_{2i-1}(t)$": sum(x(2*i - 1, t) for i in range(1, 21)),
}

plt.figure(figsize=(12, 5))
for label, y in signals.items():
    plt.plot(t, y, label=label, linewidth=1)

plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title("Sampled signals, Fs = 1000 Hz, duration = 2 s")
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()

```
