```python
import numpy as np
import matplotlib.pyplot as plt

Fs = 1000
Ts = 1.0 / Fs
duration = 1.0

n = np.arange(int(duration * Fs))
t = n * Ts

x = np.sin(20 * np.pi * t)

x_shift = np.sin(20 * np.pi * (t - 0.05))
x_reverse = np.sin(20 * np.pi * (-t))
x_amp = -2 * x
x_offset = x + 0.5

plt.figure(figsize=(10, 5))

plt.plot(t, x, label="Original")
plt.plot(t, x_shift, label="Shifted right by 0.05 s")
plt.plot(t, x_reverse, label="Time reversed")
plt.plot(t, x_amp, label="Amplitude multiplied by -2")
plt.plot(t, x_offset, label="Shifted upward by 0.5")

plt.xlabel("Time [s]")
plt.ylabel("Amplitude")
plt.title(r"Signal operations for $x(t)=\sin(20\pi t)$")
plt.grid(True)
plt.legend()
plt.show()
```
