$x[n] = \sin(\pi n / 8)$ に対して $N = 128$ 点の矩形窓をかけた信号 $x[n] w[n]$ を，
横軸を$n$ として stem プロットしなさい．また，窓関数 $w[n]$ を折れ線グラフとして重ねてプロットしなさい．
$N = 128$ 点のハン窓を用いた場合についても，窓をかけた信号と窓関数を同様にプロットしなさい．

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

figure_dir = Path("outputs/figures")
figure_dir.mkdir(parents=True, exist_ok=True)

N = 128
n = np.arange(N)
x = np.sin(np.pi * n / 8)

w_rect = np.ones(N)
x_rect = x * w_rect

w_hann = np.hanning(N)
x_hann = x * w_hann

print(f"N = {N}")
print(f"x[n] の最大値: {x.max():.3f}")
print(f"矩形窓をかけた信号の最大値: {x_rect.max():.3f}")
print(f"ハン窓をかけた信号の最大値: {x_hann.max():.3f}")
```

```python
plt.figure(figsize=(10, 4))
plt.stem(n, x_rect, basefmt=" ")
plt.plot(n, w_rect, linewidth=2)
plt.xlabel("n")
plt.ylabel("amplitude")
plt.title("Rectangular window")
plt.grid(True)
plt.tight_layout()
plt.savefig(figure_dir / "05_rectangular_windowed_signal.png", dpi=150)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.stem(n, x_hann, basefmt=" ")
plt.plot(n, w_hann, linewidth=2)
plt.xlabel("n")
plt.ylabel("amplitude")
plt.title("Hann window")
plt.grid(True)
plt.tight_layout()
plt.savefig(figure_dir / "05_hann_windowed_signal.png", dpi=150)
plt.show()
```

矩形窓では区間内の振幅がそのまま残る．
ハン窓では両端が $0$ に近づくため，窓をかけた信号の両端も滑らかに小さくなる．
