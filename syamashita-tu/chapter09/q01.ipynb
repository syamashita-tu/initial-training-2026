代表的な音律である12平均律において，
音階（ド，ド＃，レ，レ＃，ミ，ファ，ファ＃，ソ，ソ＃，ラ，ラ＃，シ）の音の周波数は，
ラ（A4）の音を $440$ Hz として隣り合う音の周波数の比が $2^{1/12}$ となるよう定められている
（1オクターブ＝12音離れると周波数が2倍になる）．
ド・レ・ミ・ファ・ソ・ラ・シの音の周波数を計算しなさい．




\begin{align}
\text{ド} &= 
\text{レ} &= 349.2282314330038
\text{ミ} &= 369.99442271163434
\text{ファ} &= 349.2282314330038
\text{ソ} &= 391.99543598174927
\text{ラ} &= 440
\text{シ} &= 466.1637615180899
\end{align}

```python
import numpy as np

a4 = 440.0
notes = {
    "ド C4": -9,
    "レ D4": -7,
    "ミ E4": -5,
    "ファ F4": -4,
    "ソ G4": -2,
    "ラ A4": 0,
    "シ B4": 2,
}

print("音名\tA4からの半音差\t周波数 [Hz]")
print("-" * 36)

frequencies = {}
for note, semitone in notes.items():
    frequency = a4 * 2 ** (semitone / 12)
    frequencies[note] = frequency
    print(f"{note}\t{semitone:>+d}\t\t{frequency:.2f}")
```
