ファに対応する周波数を持つ正弦波を
サンプリング周波数 $F_s = 16000$ Hz で 3 秒間サンプリングした信号を作り，
保存しなさい

```python
from pathlib import Path
import numpy as np
import soundfile as sf
from IPython.display import Audio, display

output_dir = Path("outputs/audio")
output_dir.mkdir(parents=True, exist_ok=True)

fs = 16000
duration = 3.0
amplitude = 0.5

f_fa = 440.0 * 2 ** (-4 / 12)

n = np.arange(int(fs * duration))
t = n / fs
x = amplitude * np.sin(2 * np.pi * f_fa * t)

output_path = output_dir / "02_sine_fa_f4.wav"
sf.write(output_path, x, fs)

print(f"ファ F4 の周波数: {f_fa:.2f} Hz")
print(f"保存先: {output_path}")
print(f"信号長: {len(x)} samples")
print(f"時間長: {len(x) / fs:.2f} s")

display(Audio(x, rate=fs))
```
