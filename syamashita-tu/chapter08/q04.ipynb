$N = 4$ のとき，DFT行列 $\boldsymbol{F}_N$ を手で求めなさい．
また， $\boldsymbol{F}_N$ を使って，$x[n] = (1, 1, 0, 0)^\top$ の DFT を手計算しなさい．
得られた $X[k]$ を， np.fft.fft を使ってDFTを計算した結果と比較しなさい．

$$
F_4 =
\begin{bmatrix}
1 & 1 & 1 & 1 \\
1 & -j & -1 & j \\
1 & -1 & 1 & -1 \\
1 & j & -1 & -j
\end{bmatrix}
$$

$$
x[n] =
\begin{bmatrix}
1 \\
1 \\
0 \\
0
\end{bmatrix}
$$

より，

$$
X[k] = F_4 x =
\begin{bmatrix}
2 \\
1-j \\
0 \\
1+j
\end{bmatrix}
$$

```python
import numpy as np

x = np.array([1, 1, 0, 0], dtype=complex)
X = np.fft.fft(x)

print(X)
```
