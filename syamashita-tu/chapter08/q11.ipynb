$N = 8192$点の信号 $x[n]$ に対して， my_dft を使って$N$点 DFT を計算するのにかかる時間を計測しなさい．
また，同じ信号に対して np.fft.fft を使って$N$点 DFT を計算するのにかかる時間を計測しなさい．

```python
import numpy as np
import time

def my_dft(x):
    x = np.asarray(x, dtype=complex)
    N = len(x)

    n = np.arange(N)
    k = n[:, None]

    F = np.exp(-1j * 2 * np.pi * k * n / N)
    return F @ x

N = 8192
n = np.arange(N)
x = np.cos(2 * np.pi * 50 * n / N)

start = time.perf_counter()
X_np = np.fft.fft(x)
time_fft = time.perf_counter() - start

print("np.fft.fft time:", time_fft)

start = time.perf_counter()
X_my = my_dft(x)
time_my = time.perf_counter() - start

print("my_dft time:", time_my)
print("max error:", np.max(np.abs(X_my - X_np)))
```

- `my_dft` は $O(N^2)$
- `np.fft.fft` は $O(N\log N)$
- $N=8192$ では `np.fft.fft` が圧倒的に高速
- 環境によっては `my_dft` がメモリ不足になる可能性あり
