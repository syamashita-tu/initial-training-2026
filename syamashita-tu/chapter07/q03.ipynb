```python
import numpy as np
import matplotlib.pyplot as plt
 
Fs = 1000           # サンプリング周波数 [Hz]
Ts = 1.0  / Fs          # サンプリング周期 [秒]
duration = 1.0        # 信号長 [秒]
 
n = np.arange(int(duration * Fs))     # 0, 1, 2, ..., N-1
t = n * Ts                            # 各サンプルに対応する時刻
x = np.sin(20 * np.pi * t)        # x[n] = sin(2π f0 (n Ts))

plt.plot(t, x, label=r"$F_s=1000$ Hz")
plt.xlabel("Time [s]"); plt.ylabel("x[n]")
plt.legend()
plt.show()
```
