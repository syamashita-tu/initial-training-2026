$x(t)=\sin{2\pi t},\quad F_s=1000\text{Hz},\quad F_s=10\text{Hz}$

```python
import numpy as np
import matplotlib.pyplot as plt
 
Fs1 = 1000
Fs2 = 10             # サンプリング周波数 [Hz]
Ts1 = 1.0  /Fs1          # サンプリング周期 [秒]
Ts2 = 1.0  /Fs2          # サンプリング周期 [秒]
duration = 1.0        # 信号長 [秒]
 
n1 = np.arange(int(duration * Fs1))     # 0, 1, 2, ..., N-1
t1 = n1 * Ts1                            # 各サンプルに対応する時刻
 
n2 = np.arange(int(duration * Fs2))     # 0, 1, 2, ..., N-1
t2 = n2 * Ts2                            # 各サンプルに対応する時

x1 = np.sin(2 * np.pi * t1)        # x[n] = sin(2π f0 (n Ts1))
x2 = np.sin(2 * np.pi * t2)        # x[n] = sin(2π f0 (n Ts2))

plt.plot(t1, x1, label=r"$F_s=1000$ Hz")
plt.stem(t2, x2, label=r"$F_s=10$ Hz")
plt.xlabel("Time [s]"); plt.ylabel("x[n]")
plt.legend()
plt.show()
```
